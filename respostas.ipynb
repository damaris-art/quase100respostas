## Exercício 32
Calcular fatorial.

```python
numero = int(input("Digite um número para calcular o fatorial: "))

fatorial = 1
conta = ""

for i in range(numero, 0, -1):
    fatorial *= i

    if i == numero:
        conta += str(i)
    else:
        conta += " x " + str(i)

print(f"{numero}! = {conta} = {fatorial}")
```

## Exercício 34
Gerar lista de primos até N.

```python
n = int(input("Digite o valor de N: "))
primos = []

for numero in range(2, n + 1):
    primo = True

    for i in range(2, numero):
        if numero % i == 0:
            primo = False

    if primo:
        primos.append(numero)

print("Primos até N:", primos)
```
